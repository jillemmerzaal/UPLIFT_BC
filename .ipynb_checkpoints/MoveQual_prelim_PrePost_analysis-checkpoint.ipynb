```python
raw_t0 = pd.read_excel('C:/Users/u0117545/Documents/GitHub/ULIFT_BC/Output/MoveQual_aff.xlsx', sheet_name = 'T0')
raw_t1 = pd.read_excel('C:/Users/u0117545/Documents/GitHub/ULIFT_BC/Output/MoveQual_aff.xlsx', sheet_name = 'T1')

raw = pd.concat([raw_t0, raw_t1], axis=0, ignore_index=True)
raw.head()
```

```python
df = raw.iloc[0:,1:]
df.index = raw.iloc[0:,0]
df.head()
```

```python
df = df.drop('BC_022',0)
```

```python
X = df.drop(['trial','time'], axis = 1)
# find minimium required rows given we want 5 columns
ncols = 4
nrows = len(X.columns) // ncols + (len(X.columns) % ncols > 0)
dx = "time"; ort = "v"; pal = 'colorblind'; sigma = .6; viol = .6; mv = .0

f = plt.figure(figsize=(20, 20))
plt.subplots_adjust(hspace=0.5)
plt.suptitle("Movement quality for UL", fontsize=18, y=0.95)

f.tight_layout();
offset = 0.1;

for n, columns in enumerate(X.columns):
    # add a new subplot iteratively
    ax = plt.subplot(nrows, ncols, n + 1)
    
    # filter df and plot ticker on the new subplot axis
    #df[df["ticker"] == ticker].plot(ax=ax)
     
    pt.RainCloud(x = dx, y = columns, data = df, ax = ax,
                palette = pal, bw = sigma, width_viol = viol,
                orient = ort, move = mv)
    sns.despine(offset=10)
    ax.set_xlabel('')
```

Berekening sample entropy:
* tollerantie op basis van het signaal r*std(signaal) --> zeer signaal specifiek

* tollerantie met een vaste waarde r = 0.2 --> mogelijk grotere kans om verschillen te vinden

* tollerantie bepaald op pre-op data en dan diezelfde tollerantie gebruiken voor de pre en post op metingen. --> individueel maar mogelijk toch makkelijker verschil tussen pre en post op te vinden!

# Statistical anlysis
### Pre-op vs Post-op on 20 participants as a first analysis method

```python
# find outliers
```

```python
T0 = df.loc[df['time']=='T0']
T1 = df.loc[df['time']=='T1']
timepoints = ['T0', 'T1']

for var in df.columns:
    print('Analyzing {}'.format(var))
    if var == 'trial' or var == 'time':
        print(80* "_" + "\n" )
    else:
        t_statistic, pvalue = stats.wilcoxon(T0[var], T1[var])
        print(t_statistic, round(pvalue, ndigits=4))
        
        print(80* "_" + "\n" )
                           
```

In general, there appears to be no effect of surgery on movement quality parameters on group level with N=20

## Quick Dash score
### 1) Split group on disabled group and non-disabled
### 2) Correlate the Quick Dash scores with movement quality. i.e. do people with a higher score perform better?

Berekening Quick DASH beperkingen en symptomen score:  
{(Som van de antwoorden op N items / N ) – 1} x 25 =      (zie refentie 5)    
N = aantal ingevulde antwoorden  
Let op: de Quick DASH score mag niet berekend worden als meer dan 1 vraag niet beantwoord is.

```python
dash = pd.read_csv("C:\\Users\\u0117545\\KU Leuven\\An De Groef - DATA\\XsensTest\\QuickDASH\\UPLIFTBC-QuickDASH_DATA_2022-10-31_1406 (1).csv",
                   sep=';', index_col=0)
dash.head()
```

```python
dash_t0 = dash[dash['redcap_event_name'].str.contains('t0')]
dash_t1 = dash[dash['redcap_event_name'].str.contains('t1')]
```

```python
som = dash_t0.iloc[:,1:].sum(axis='columns')
col_n_counts = np.sum(dash_t0.applymap(lambda x: 0 if pd.isnull(x) else 1), axis=1)-1
complete = col_n_counts.apply(lambda x:  x >= 10 )

total_score_t0 = ((som/col_n_counts)-1) * 25
```

```python
t0_dash_total = pd.DataFrame(total_score_t0, columns=['Score'])
t0_dash_total['complete'] = complete
t0_dash_total
```

```python
som = dash_t1.iloc[:,1:].sum(axis='columns')
col_n_counts = np.sum(dash_t1.applymap(lambda x: 0 if pd.isnull(x) else 1), axis=1)-1
complete = col_n_counts.apply(lambda x: x >= 10)

total_score_t1 = ((som/col_n_counts)-1) * 25
```

```python
t1_dash_total = pd.DataFrame(total_score_t1, columns=['Score'])
t1_dash_total['complete'] = complete
t1_dash_total
```
