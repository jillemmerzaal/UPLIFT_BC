```python
import numpy as np
import pandas as pd
import pandas as pd
import spm1d
import os
import matplotlib.pyplot as plt
```

```python

jointsOfInterst =['Scapula_abbuction',
    'Scapula_rotation',
    'Scapula_flexion',
    'Glenohumeraal_abbuction',
    'Glenohumeraal_rotation',
    'Glenohumeraal_flexion',
    'Elbow_flexion',
    'Trunk_lateroflexion',
    'Trunk_rotation',
    'Trunk_flexion']

#jointsOfInterst = ['Scapula_abbuction', 'Glenohumeraal_flexion']
```

# Independent t-test

for jnt in jointsOfInterst:
    df = pd.read_excel("C:\\Users\\u0117545\\Documents\\GitHub\\ULIFT_BC\\Phase4.xlsx", sheet_name =jnt)
    
    print("Before removing rows with missing data: {} rows".format(len(df)))    
    no_missing_data = (~df.isnull().any(axis=1))
    df = df.loc[no_missing_data]
    print("After removing rows with missing data: {} rows".format(len(df)))

    Y0 = df.loc[df['Time'] == 'T0'].iloc[:,2:]
    Y1 = df.loc[df['Time'] == 'T1'].iloc[:,2:]


    f, ax = plt.subplots(nrows =2, ncols=1)

    spm1d.plot.plot_mean_sd(Y0, ax[0], linecolor='b', facecolor=(0.7,0.7,1), edgecolor='b', label='preOP')
    spm1d.plot.plot_mean_sd(Y1, ax[0], linecolor='r', facecolor=(1,0.7,0.7), edgecolor='r', label='postOP')
    

    t  = spm1d.stats.ttest2(Y0, Y1, equal_var=False)
    ti = t.inference(alpha=0.05, two_tailed=True)
    ax[1].plot(ti.z, color = 'k')
    [ax[1].axhline(y=i, linestyle='--', color='k') for i in [ti.zstar*-1, 0,  ti.zstar]]
    
    
    plt.xlim([0, 100])
    ax[0].set_xlabel('%')
    ax[0].set_ylabel('joint Angle deg')
    f.suptitle(jnt)
    plt.show()

# Paired t-test

```python
for jnt in jointsOfInterst:
    df = pd.read_excel("C:\\Users\\u0117545\\Documents\\GitHub\\ULIFT_BC\\ULIFT\\Phase1.xlsx", sheet_name =jnt)
    
    print("Before removing rows with missing data: {} rows".format(len(df)))    
    no_missing_data = (~df.isnull().any(axis=1))
    df = df.loc[no_missing_data]
    
    # remove BC_014 due to fautly post-op data
    i = df[df['ppID'] == 'BC_014'].index
    df = df.drop(i)
    print("After removing rows with missing data: {} rows".format(len(df)))


    # define data for pre-op and 6wks post-op
    Y0 = df.loc[df['Time'] == 'T0'].iloc[:,2:]
    Y1 = df.loc[df['Time'] == 'T1'].iloc[:,2:]


    f, ax = plt.subplots(nrows =2, ncols=1)

    spm1d.plot.plot_mean_sd(Y0, ax[0], linecolor='b', facecolor=(0.7,0.7,1), edgecolor='b', label='preOP')
    spm1d.plot.plot_mean_sd(Y1, ax[0], linecolor='r', facecolor=(1,0.7,0.7), edgecolor='r', label='postOP')
    

    t  = spm1d.stats.ttest_paired(Y0, Y1)
    ti = t.inference(alpha=0.05)
    ax[1].plot(ti.z, color = 'k')
    [ax[1].axhline(y=i, linestyle='--', color='k') for i in [ti.zstar*-1, 0,  ti.zstar]]
    
    
    plt.xlim([0, 100])
    ax[0].set_xlabel('%')
    ax[0].set_ylabel('joint Angle deg')
    f.suptitle(jnt)
    plt.show()
```
